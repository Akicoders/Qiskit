```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_city
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

state = Statevector.from_label("+-rl")
circ_3 = QuantumCircuit(4)
circ_3.initialize(state.data, [0, 1, 2, 3])

# Aplicar las puertas cunticas
circ_3.x(0)
circ_3.cx(1, 2)
circ_3.h(1)
circ_3.cx(0, 1)
circ_3.x(0)
circ_3.z(1)
circ_3.x(2)

# Dibujar el circuito
circ_3.draw(output='mpl')

final_state = Statevector.from_instruction(circ_3)
plot_state_city(final_state, title="Estado final")
plt.show()
```
