```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import circuit_drawer, plot_histogram
```

```python
circ_02 = QuantumCircuit(2,2)
circ_02.x(1)
circ_02.cx(0,1)
circ_02.cx(1,0)
circ_02.cx(0,1)
circ_02.swap(0,1)
circ_02.draw(output="mpl")
```

```python
state = Statevector.from_int(0, 2**2)
state = state.evolve(circ_02) 
state.draw('latex')
```
