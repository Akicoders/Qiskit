# Algoritmo Deustch-Jozsa
<p> Es un algoritmo para saber si la funcion es balanceada o constante  </p>


```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer, AerSimulator, QasmSimulator
from qiskit.visualization import circuit_drawer, plot_histogram
import pylatexenc
```

```python
circl = QuantumCircuit(4, 3)
circl.x(3)
circl.h((0, 1, 2, 3))
circuit_drawer(circl, output="mpl", style={"backgroundcolor": "#EEEEEE"})
```

```python
circ2 = QuantumCircuit(4, 3)
circ2.barrier(range(4))
circ2.cx(0,3)
circ2.x(0)
circ2.cx(0,3)
circ2.cx(1,3)
circ2.x(1)
circ2.cx(1,3)
circ2.cx(2,3)
circ2.x(2)
circ2.cx(2,3)
circ2.barrier(range(4))
circ2.draw(output="mpl", style={"backgroundcolor": "#EEEEEE"})
```

```python
circ3 = QuantumCircuit(4, 3)

circ3.h((0, 1, 2))
circ3.measure((0, 1, 2), (0, 1, 2))
circ3.draw(output="mpl")
```
